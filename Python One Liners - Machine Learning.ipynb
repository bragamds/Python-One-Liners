### Linear Regression

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
apple = np.array([155,156,157])
n = len(apple)
```

```python
model = LinearRegression().fit(np.arange(n).reshape((n,1)), apple)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
x = np.array([[0,'No'],
             [10, 'No'],
             [60, 'Yes'],
             [90, 'Yes']])
```

```python
model = LogisticRegression().fit(x[:,0:1].reshape(4,1), x[:,1])
print(model.predict([[2],[12],[13],[40],[90]]))
```

```python
for i in range(20):
    print("x=" + str(i) + "--->" + str(model.predict_proba([[i]])))
```

### K-Means Clustering in One Line

Supervised Learning: 

Labeled Training Data - You know the output value of every input value in the training data. 


Unsupervised Learning:

Unlabeled Training Data - No output to start with. Prediction impossible. Can still get useful knowledge from the unlabelled dataset. 

Step 1: Choose the number K of clusters

Step 2: Select random K points as cluster centres

> e.g. Choose 5 random points

Step 3: Assign each data point to the nearest cluster

Step 4: Compute and place the new centroid of each cluster

The star shows the initial position of the cluster. Notice how some of the observations have changed clusters. Almost all observations in the middle are connected to the purple cluster now.

Step 5: Repeat step 4 until no observations change cluster

Finally, we just repeat step 4 until no observations change cluster and we have our final clusters specified. For the graph below we are just on the second iteration, and we see that the clusters already look fairly reasonable. In reality we might have to repeat step 4 one or two times more before the clusters settle to their real centers.

https://towardsdatascience.com/visual-introduction-to-clustering-with-kmeans-89e71e97c60f

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
X = np.array([[35, 7000], [45, 6900], [70, 7100],
 [20, 2000], [25, 2200], [15, 1800]])
```

```python
cluster = KMeans(n_clusters = 2).fit(X)
```

```python
cc = cluster.cluster_centers_
```

```python
cc
```

```python
xaxis = X[:,0]
yaxis = X[:,1]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plot1 = sns.scatterplot(x=xaxis, y=yaxis, facecolor = 'green')
plot2 = sns.scatterplot(x=cc[:,0], y = cc[:,1])
plot1.set_xlabel('Hours Worked')
plot1.set_ylabel('Salary')
```

```python
plt.scatter(x=xaxis, y=yaxis);
```

### K-Nearest Neighbors in One Line

KNN uses the whole dataset as a model

Can use for prediction or classification

KNN for regression (need to use KNeighborsRegressor)

KNN for classification (need to use KNeighborsClassifier)

The idea in kNN methods is to identify ‘k’ samples in the dataset that are similar or close in the space. Then we use these ‘k’ samples to estimate the value of the missing data points. Each sample’s missing values are imputed using the mean value of the ‘k’-neighbors found in the dataset.

```python
## Dependencies
from sklearn.neighbors import KNeighborsRegressor
import numpy as np

## Data (House Size (square meters) / House Price ($))
X = np.array([[35, 30000], [45, 45000], [40, 50000],
 [35, 35000], [25, 32500], [40, 40000]])
```

```python
KNN = KNeighborsRegressor(n_neighbors=3)
KNN = KNN.fit(X[:,0].reshape(-1,1), X[:,1])
```

```python
res = KNN.predict([[30]])
print(res)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x=X[:,0], y=X[:,1])
```

### Classification with Support-Vector Machines in One Line

Example of Machine Learning Classification


```python
from sklearn import svm
import numpy as np
```

```python
X = np.array([[9, 5, 6, "computer science"],
 [10, 1, 2, "computer science"],
 [1, 8, 1, "literature"],
 [4, 9, 3, "literature"],
 [0, 1, 10, "art"],
 [5, 7, 9, "art"]])
```

```python
svm = svm.SVC().fit(X[:,0:3],X[:,-1])
```

```python
student_0 = svm.predict([[3,3,6]])
student_1 = svm.predict([[8,1,1]])
```

```python
print(student_0, student_1)
```
